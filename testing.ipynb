```python
import aerosandbox as asb
import aerosandbox.numpy as np
import matplotlib.pyplot as plt
import aerosandbox.tools.pretty_plots as p
```

Importación inicial

```python
coordinate_airfoil = asb.Airfoil("dae11")
```

Convertir a Kulfan

```python
kulfan_airfoil = coordinate_airfoil.to_kulfan_airfoil()
kulfan_airfoil.kulfan_parameters
```

### Definir un airfoil con Kulfan

```python
# Define los parámetros como variables
lower_weights = -0.2 * np.ones(15)
upper_weights = 0.3 * np.ones(8)
leading_edge_weight = 0.2
TE_thickness = 0.005

lower_weights[10] = 0.6
lower_weights[13] = 0.6

print(lower_weights)

# Crea el perfil con Kulfan utilizando las variables como parámetros
paramairfoil = asb.KulfanAirfoil(
    name="My Airfoil",
    lower_weights=lower_weights,
    upper_weights=upper_weights,
    leading_edge_weight=leading_edge_weight,
    TE_thickness=TE_thickness
)


```

```python
fig, ax = plt.subplots(figsize=(6, 2))
paramairfoil.draw()
```
